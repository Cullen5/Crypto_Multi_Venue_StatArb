```python
# --- Top 25% vs Bottom 25% pairs: average pair-level stats ---
n = len(df)
q = n // 4

top_25 = df.head(q)
bot_25 = df.tail(q)

numeric_cols = df.select_dtypes(include="number").columns.tolist()

comparison = pd.DataFrame({
    "Top 25%": top_25[numeric_cols].mean(),
    "Bottom 25%": bot_25[numeric_cols].mean(),
})
comparison["Delta"] = comparison["Top 25%"] - comparison["Bottom 25%"]
comparison
```

```python
pd.set_option("display.float_format", lambda x: f"{x:.6f}")
# --- Top 25% vs Bottom 25% pairs: average individual trade stats ---
top_pair_ids = set(top_25["pair_id"])
bot_pair_ids = set(bot_25["pair_id"])

top_trades = df_trades[df_trades["pair_id"].isin(top_pair_ids)]
bot_trades = df_trades[df_trades["pair_id"].isin(bot_pair_ids)]

trade_numeric = df_trades.select_dtypes(include="number").columns.tolist()

trade_comparison = pd.DataFrame({
    "Top 25%": top_trades[trade_numeric].mean(),
    "Bottom 25%": bot_trades[trade_numeric].mean(),
})
trade_comparison["Delta"] = trade_comparison["Top 25%"] - trade_comparison["Bottom 25%"]

print(f"Top 25% trades: {len(top_trades)}  |  Bottom 25% trades: {len(bot_trades)}")
trade_comparison
```

```python
# --- Top 25% vs Bottom 25% of individual trades by P/L ---
q75 = df_trades["gross_pnl"].quantile(0.75)
q25 = df_trades["gross_pnl"].quantile(0.25)

top_trades = df_trades[df_trades["gross_pnl"] >= q75]
bot_trades = df_trades[df_trades["gross_pnl"] <= q25]

trade_numeric = df_trades.select_dtypes(include="number").columns.tolist()

trade_comparison = pd.DataFrame({
    "Top 25% (Winners)": top_trades[trade_numeric].mean(),
    "Bottom 25% (Losers)": bot_trades[trade_numeric].mean(),
})
trade_comparison["Delta"] = trade_comparison["Top 25% (Winners)"] - trade_comparison["Bottom 25% (Losers)"]

print(f"Top 25% trades: {len(top_trades)}  |  Bottom 25% trades: {len(bot_trades)}")
trade_comparison
```

```python
# === Full Analysis: Quick Trades (1-3 Bars Held) ===
quick_trades = df_trades[df_trades["bars_held"].between(1, 3)]
other_trades = df_trades[~df_trades["bars_held"].between(1, 3)]

print(f"Quick trades: {len(quick_trades)} ({len(quick_trades)/len(df_trades)*100:.1f}% of all trades)")
print(f"Other trades: {len(other_trades)}")
print()

# --- 1) Basic P/L comparison ---
compare = pd.DataFrame({
    "Quick (1-3 bars)": quick_trades.select_dtypes("number").mean(),
    "All Others":       other_trades.select_dtypes("number").mean(),
})
compare["Delta"] = compare["Quick (1-3 bars)"] - compare["All Others"]
print("=== Avg Stats: Quick vs Others ===")
display(compare)

# --- 2) Breakdown by bars_held (1, 2, 3 separately) ---
print("\n=== Breakdown by Exact Bars Held ===")
per_bar = quick_trades.groupby("bars_held").agg(
    count=("net_pnl", "size"),
    avg_pnl=("net_pnl", "mean"),
    median_pnl=("net_pnl", "median"),
    total_pnl=("net_pnl", "sum"),
    win_rate=("net_pnl", lambda x: (x > 0).mean()),
    avg_fees=("total_cost", "mean"),          # adjust col name
)
display(per_bar)

# --- 3) Win rate & avg win vs avg loss ---
winners = quick_trades[quick_trades["net_pnl"] > 0]
losers  = quick_trades[quick_trades["net_pnl"] <= 0]
print(f"\n=== Win/Loss Profile ===")
print(f"Win rate:     {len(winners)/len(quick_trades)*100:.1f}%")
print(f"Avg winner:   {winners['net_pnl'].mean():.4f}")
print(f"Avg loser:    {losers['net_pnl'].mean():.4f}")
print(f"Profit factor: {winners['net_pnl'].sum() / abs(losers['net_pnl'].sum()):.2f}")
print(f"Avg fees (quick): {quick_trades['total_cost'].mean():.4f}")
print(f"Avg fees (other): {other_trades['total_cost'].mean():.4f}")

# --- 4) Fees as % of gross move ---
# Are fees eating the entire move on short-duration trades?
if "gross_pnl" in quick_trades.columns:
    quick_trades["fee_pct_of_gross"] = quick_trades["total_cost"].abs() / quick_trades["gross_pnl"].abs()
    print(f"\n=== Fee Impact ===")
    print(f"Median fees as % of gross move (quick): {quick_trades['fee_pct_of_gross'].median()*100:.1f}%")
    other_trades["fee_pct_of_gross"] = other_trades["total_cost"].abs() / other_trades["gross_pnl"].abs()
    print(f"Median fees as % of gross move (other): {other_trades['fee_pct_of_gross'].median()*100:.1f}%")

# --- 5) Entry confidence distribution ---
print("\n=== Entry Confidence ===")
print(f"Avg confidence (quick): {quick_trades['entry_confidence'].mean():.3f}")
print(f"Avg confidence (other): {other_trades['entry_confidence'].mean():.3f}")

# --- 6) Which pairs dominate quick trades? ---
print("\n=== Top 10 Pairs by Quick Trade Count ===")
pair_breakdown = (
    quick_trades.groupby("pair_id")
    .agg(count=("net_pnl", "size"), avg_pnl=("net_pnl", "mean"), total_pnl=("net_pnl", "sum"))
    .sort_values("count", ascending=False)
    .head(10)
)
display(pair_breakdown)

# --- 7) Exit reason distribution ---
if "exit_reason" in quick_trades.columns:
    print("\n=== Exit Reasons (Quick vs Others) ===")
    quick_exits = quick_trades["exit_reason"].value_counts(normalize=True).rename("Quick %")
    other_exits = other_trades["exit_reason"].value_counts(normalize=True).rename("Other %")
    exit_compare = pd.concat([quick_exits, other_exits], axis=1).fillna(0)
    display(exit_compare)

# --- 8) Visualization ---
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# P/L distribution
axes[0, 0].hist(quick_trades["net_pnl"], bins=50, alpha=0.6, color="red", label="Quick")
axes[0, 0].hist(other_trades["net_pnl"], bins=50, alpha=0.4, color="blue", label="Other")
axes[0, 0].axvline(0, color="black", linestyle="--")
axes[0, 0].set_title("P/L Distribution: Quick vs Others")
axes[0, 0].legend()

# P/L by bars held (box plot)
df_trades.boxplot(column="net_pnl", by="bars_held", ax=axes[0, 1])
axes[0, 1].set_title("P/L by Bars Held")
axes[0, 1].axhline(0, color="red", linestyle="--")
axes[0, 1].set_xlabel("Bars Held")

# Entry confidence: quick vs other
axes[1, 0].hist(quick_trades["entry_confidence"], bins=30, alpha=0.6, color="red", label="Quick")
axes[1, 0].hist(other_trades["entry_confidence"], bins=30, alpha=0.4, color="blue", label="Other")
axes[1, 0].set_title("Entry Confidence Distribution")
axes[1, 0].legend()

# Cumulative P/L of quick trades over time
quick_sorted = quick_trades.sort_values("entry_bar")
axes[1, 1].plot(quick_sorted["entry_bar"], quick_sorted["net_pnl"].cumsum(), color="red")
axes[1, 1].axhline(0, color="black", linestyle="--")
axes[1, 1].set_title("Cumulative P/L of Quick Trades")

plt.suptitle("Quick Trades (1-3 Bars) Deep Dive", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Crisis period analysis
crisis_periods = {
    "UST/Luna Collapse": ("2022-05-01", "2022-06-15"),
    "FTX Bankruptcy": ("2022-11-01", "2022-12-15"),
    "USDC Depeg / Bank Crisis": ("2023-03-01", "2023-04-01"),
    "SEC Lawsuits (Binance/Coinbase)": ("2023-06-01", "2023-07-15"),
}

# need timestamps on trades  map entry_bar to timestamp from equity curve
eq_df = results.equity_curve
bar_to_ts = dict(zip(eq_df["bar"], pd.to_datetime(eq_df["timestamp"])))
df_trades["entry_ts"] = df_trades["entry_bar"].map(bar_to_ts)
df_trades["exit_ts"] = df_trades["exit_bar"].map(bar_to_ts)

# per-crisis stats
crisis_stats = []
for name, (start, end) in crisis_periods.items():
    start_ts, end_ts = pd.Timestamp(start), pd.Timestamp(end)

    # trades that overlapped with the crisis (open during any part of it)
    mask = (
        (df_trades["entry_ts"] <= end_ts) &
        (df_trades["exit_ts"] >= start_ts)
    )
    ct = df_trades[mask]

    if len(ct) == 0:
        crisis_stats.append({"Crisis": name, "Trades": 0})
        continue

    wins = ct[ct["net_pnl"] > 0]
    losses = ct[ct["net_pnl"] <= 0]

    crisis_stats.append({
        "Crisis": name,
        "Period": f"{start}  {end}",
        "Trades": len(ct),
        "Win Rate": f"{len(wins)/len(ct):.1%}",
        "Total P/L": f"${ct['net_pnl'].sum():,.2f}",
        "Avg P/L %": f"{ct['pnl_pct'].mean()*100:.3f}%",
        "Avg P/L $": f"${ct['net_pnl'].mean():,.2f}",
        "Best Trade": f"${ct['net_pnl'].max():,.2f}",
        "Worst Trade": f"${ct['net_pnl'].min():,.2f}",
        "Total Costs": f"${ct['total_cost'].sum():,.2f}",
        "Avg Bars Held": f"{ct['bars_held'].mean():.0f}",
        "Exit Reasons": dict(ct["exit_reason"].value_counts()),
    })

crisis_df = pd.DataFrame(crisis_stats)
print("\n=== CRISIS PERIOD PERFORMANCE ===\n")
for _, row in crisis_df.iterrows():
    print(f"  {row['Crisis']}")
    if row["Trades"] == 0:
        print("    No trades during this period\n")
        continue
    print(f"    Period:       {row['Period']}")
    print(f"    Trades:       {row['Trades']}  (WR: {row['Win Rate']})")
    print(f"    Total P/L:    {row['Total P/L']}  |  Avg: {row['Avg P/L $']}  ({row['Avg P/L %']})")
    print(f"    Best/Worst:   {row['Best Trade']} / {row['Worst Trade']}")
    print(f"    Costs:        {row['Total Costs']}")
    print(f"    Avg Hold:     {row['Avg Bars Held']} bars")
    print(f"    Exits:        {row['Exit Reasons']}")
    print()

# chart: P/L by crisis
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

names = [s["Crisis"] for s in crisis_stats if s["Trades"] > 0]
totals = [df_trades[
    (df_trades["entry_ts"] <= pd.Timestamp(crisis_periods[n][1])) &
    (df_trades["exit_ts"] >= pd.Timestamp(crisis_periods[n][0]))
]["net_pnl"].sum() for n in names]
colors = ["#2ca02c" if v >= 0 else "#d62728" for v in totals]

axes[0].barh(names, totals, color=colors, alpha=0.7)
axes[0].axvline(0, color="black", linestyle="--", alpha=0.4)
axes[0].set_xlabel("Total P/L ($)")
axes[0].set_title("Total P/L by Crisis")

win_rates = [df_trades[
    (df_trades["entry_ts"] <= pd.Timestamp(crisis_periods[n][1])) &
    (df_trades["exit_ts"] >= pd.Timestamp(crisis_periods[n][0]))
]["net_pnl"].apply(lambda x: x > 0).mean() for n in names]

axes[1].barh(names, win_rates, color="#1f77b4", alpha=0.7)
axes[1].axvline(0.5, color="black", linestyle="--", alpha=0.4)
axes[1].set_xlabel("Win Rate")
axes[1].set_title("Win Rate by Crisis")
axes[1].set_xlim(0, 1)

plt.tight_layout()
plt.show()

# sector breakdown per crisis
print("\n=== SECTOR BREAKDOWN PER CRISIS ===\n")
for name, (start, end) in crisis_periods.items():
    start_ts, end_ts = pd.Timestamp(start), pd.Timestamp(end)
    mask = (df_trades["entry_ts"] <= end_ts) & (df_trades["exit_ts"] >= start_ts)
    ct = df_trades[mask]
    if len(ct) == 0:
        continue
    print(f"  {name}:")
    sector_pnl = ct.groupby("sector")["net_pnl"].agg(["sum", "count", "mean"])
    sector_pnl.columns = ["Total P/L", "Trades", "Avg P/L"]
    sector_pnl = sector_pnl.sort_values("Total P/L", ascending=False)
    with pd.option_context("display.float_format", "{:.2f}".format):
        print(sector_pnl.to_string(index=True))
    print()
```
